```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from kmeans import kmeans
```

generating random 2D data points using make_blobs 


```python
k=5
sigma =0.6
n =200
x,y = make_blobs(n_samples=n,centers=k,random_state=0,cluster_std=sigma)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

"random" initialisation of the centroids 


```python
model1 = kmeans(5,"random",100)
model1.fit(x)

predicted_Y1 = model1.predict(x)

plt.scatter(x[:,0],x[:,1],c=predicted_Y1)
```

"kmeans++" initialisation :



```python
model2 = kmeans(5,"kmeans++",100)
model2.fit(x)
predicted_Y2 = model2.predict(x)
plt.scatter(x[:,0],x[:,1],c=predicted_Y2)
```

conclusion :
the algorithms performs better when the centroids are initialised using "kmeans++"
